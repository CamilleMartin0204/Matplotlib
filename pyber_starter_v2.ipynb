```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
city_data_df = city_data_df.drop_duplicates('city')

# Combine the data into a single dataset
pyber_data = pd.merge(city_data_df, ride_data_df, on=["city","city"],how="left")

# Display the data table for preview
pyber_data.head()
```

# Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_city_data = pyber_data[pyber_data["type"]== "Urban"]
rural_city_data = pyber_data[pyber_data["type"]== "Rural"]
suburban_data = pyber_data[pyber_data["type"]== "Suburban"]

# Build the scatter plots for each city types
urban_ride_count =urban_city_data.groupby(["city"]).count()["ride_id"]
#print(urban_ride_count)
urban_average_fair =urban_city_data.groupby(["city"]).mean()["fare"]
#print(urban_average_fair)

# Incorporate the other graph properties
urban_average_driver_count =urban_city_data.groupby(["city"]).mean()["driver_count"]
#print(urban_average_driver_count)

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
plt.scatter(urban_ride_count, urban_average_fair, s=10*urban_average_driver_count,marker="o",label="urban",c="pink", linewidths=1, edgecolor="black")
# s=scale
```

```python
# Show plot
plt.show()
```
